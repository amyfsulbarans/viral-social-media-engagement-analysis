# Viral Content & Social Media Engagement Analysis

## Understanding Platform Performance and Content Effectiveness

## Project Overview

This project analyzes 5,000 viral social media posts across TikTok, Instagram, Twitter, and YouTube to identify engagement patterns and content performance.

The goal is to understand which platforms and content types drive the highest engagement and provide actionable recommendations for marketing strategies.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.read_csv("Viral_Social_Media_Trends.csv")
df.head()
```

## Data Cleaning

Column names were standardized and key engagement metrics were created.

```python
df.shape
df.columns
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
df["engagement_total"] = df["likes"] + df["shares"] + df["comments"]

df["engagement_rate"] = df["engagement_total"] / df["views"]
```

## Platform Performance Analysis

This analysis evaluates which social media platforms generate the highest engagement rates.

```python
df.groupby("platform")["engagement_rate"].mean().sort_values(ascending=False)
```

### Key Insight

Instagram shows the highest engagement rate among all platforms, indicating stronger audience interaction and content effectiveness.

TikTok also performs strongly, suggesting that short-form and visually engaging content drives higher user engagement.

## Content Type Performance

This analysis explores which types of content generate the highest engagement.

```python
df.groupby("content_type")["engagement_rate"].mean().sort_values(ascending=False)
```

### Key Insight

Short-form content (Shorts) significantly outperforms other formats in terms of engagement rate, highlighting the growing importance of concise, high-impact content in social media strategies.

## Visualization

```python
df.groupby("platform")["engagement_rate"].mean().sort_values().plot(kind="barh")
plt.title("Average Engagement Rate by Platform")
plt.xlabel("Engagement Rate")
plt.ylabel("Platform")
plt.show()
```

## Final Conclusion

The analysis reveals that both platform selection and content format play a critical role in driving engagement.

Instagram and TikTok lead in performance, while short-form content such as Shorts delivers the highest engagement rates.

These findings suggest that brands should prioritize short-form video content and focus distribution on high-performing platforms to maximize audience interaction.

## Recommendations

- Focus on Instagram and TikTok for content distribution  
- Prioritize short-form video formats (Shorts)  
- Optimize content for quick consumption and high interaction  
- Test live content strategies for additional engagement
